```python
# Gender Distribution

plt.figure(figsize=(6,5))

ax = sns.countplot(
    data=df,
    x="Gender",
    palette="Set2"
)

# Add Count Labels
for container in ax.containers:
    ax.bar_label(container)

plt.title("Customer Gender Distribution", fontsize=15, fontweight="bold")

plt.xlabel("Gender", fontsize=12)

plt.ylabel("Number of Customers", fontsize=12)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"],bins=15,kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
category_sales = (
    df.groupby("Product Category")["Total Amount"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=category_sales.index,
    y=category_sales.values,
    palette="viridis"
)

# Add Values on Bars
for i, value in enumerate(category_sales.values):
    ax.text(i, value, f"{value:.0f}",
            ha="center",
            va="bottom",
            fontsize=10)

plt.title("Total Sales by Product Category",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Product Category")

plt.ylabel("Total Sales")

plt.tight_layout()

plt.show()
```

```python
quantity = (
    df.groupby("Product Category")["Quantity"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=quantity.index,
    y=quantity.values,
    palette="magma"
)

for i, value in enumerate(quantity.values):
    ax.text(i, value, str(value),
            ha="center",
            va="bottom")

plt.title("Quantity Sold by Product Category",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Product Category")

plt.ylabel("Quantity Sold")

plt.tight_layout()

plt.show()
```

```python
# Month Order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

monthly_sales = (
    df.groupby("Month")["Total Amount"]
      .sum()
      .reindex(month_order)
)

plt.figure(figsize=(12,5))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker="o",
    linewidth=2.5
)

# Add Value Labels
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y, f"{int(y)}",
             ha="center",
             va="bottom",
             fontsize=9)

plt.title("Monthly Sales Trend (2023)",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Month")

plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.grid(alpha=0.1)

plt.tight_layout()

plt.show()
```

```python
monthly_sales.to_frame(name="Total Sales")
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))

correlation = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation Matrix",
          fontsize=15,
          fontweight="bold")

plt.tight_layout()

plt.show()
```

```python
# Total Amount Distribution

plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Total Amount",
    bins=20,
    kde=True,
    color="royalblue"
)

plt.title("Distribution of Total Sales Amount",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Total Amount")

plt.ylabel("Frequency")

plt.tight_layout()

plt.show()
```

```python
# Box Plot

plt.figure(figsize=(8,2.5))

sns.boxplot(
    x=df["Total Amount"],
    color="orange"
)

plt.title("Box Plot of Total Sales Amount",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Total Amount")

plt.tight_layout()

plt.show()
```

```python
# Revenue by Gender

gender_sales = (
    df.groupby("Gender")["Total Amount"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(6,5))

ax = sns.barplot(
    x=gender_sales.index,
    y=gender_sales.values,
    palette="Set1"
)

for i, value in enumerate(gender_sales.values):
    ax.text(
        i,
        value,
        f"{value:.0f}",
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.title("Revenue Generated by Gender",
          fontsize=15,
          fontweight="bold")

plt.xlabel("Gender")

plt.ylabel("Total Revenue")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="Product Category",
    hue="Gender",
    palette="Set2"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title(
    "Product Category Purchases by Gender",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Product Category")

plt.ylabel("Number of Transactions")

plt.tight_layout()

plt.show()
```

```python
# Product Category-wise Revenue

category_revenue = (
    df.groupby("Product Category")["Total Amount"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(5,5))

plt.pie(
    category_revenue,
    labels=category_revenue.index,
    autopct="%1.1f%%",
    startangle=90,
    explode=[0.03]*len(category_revenue),
    shadow=True
)

plt.title(
    "Product Category-wise Revenue Contribution",
    fontsize=15,
    fontweight="bold"
)

plt.tight_layout()

plt.show()
```

```python
# Month Order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

monthly_transactions = (
    df.groupby("Month")["Transaction ID"]
      .count()
      .reindex(month_order)
)

plt.figure(figsize=(12,5))

ax = sns.barplot(
    x=monthly_transactions.index,
    y=monthly_transactions.values,
    palette="viridis"
)

# Add labels on bars
for container in ax.containers:
    ax.bar_label(container)

plt.title(
    "Monthly Number of Transactions",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Month")

plt.ylabel("Number of Transactions")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# Create Age Groups

bins = [0, 20, 30, 40, 50, 60, 100]

labels = ["<20", "21-30", "31-40", "41-50", "51-60", "60+"]

df["Age Group"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels
)
```

```python
age_sales = (
    df.groupby("Age Group")["Total Amount"]
      .sum()
)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=age_sales.index,
    y=age_sales.values,
    palette="Blues_r"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title(
    "Revenue by Age Group",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Age Group")

plt.ylabel("Total Revenue")

plt.tight_layout()

plt.show()
```
